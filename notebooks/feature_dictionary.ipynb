# Feature Dictionary

*Dictionary outlining all features availables after the feature engineering pipeline has been executed. All features are subject to filtering based on availability 2 hours prior to scheduled departure.*

## Categorical

- **TAIL_NUM** — unique identifier of a plane  
- **YEAR** — year of flight (2015–2019)  
- **QUARTER** — quarter of the year (1–4)  
- **MONTH** — month of the year (1=January … 12=December)  
- **DAY_OF_WEEK** — day of the week (1=Monday … 7=Sunday)  
- **DAY_OF_MONTH** — day of the month (1–31)  
- **OP_UNIQUE_CARRIER** — airline code  
- **ORIGIN** — 3-letter origin airport code  
- **DEST** — 3-letter destination airport code  
- **PREV_FLIGHT_FLAG** — previous flight found on same tail number on the same/previous calendar day (binary)  
- **PREV2_FLIGHT_FLAG** — same as above but checks two previous flights (binary)  
- **NEXT_FLIGHT_FLAG** — next flight found on same tail number on the same/next calendar day (binary)  
- **HOLIDAY_WITHIN_3D** — flight within 3 days of a national holiday (binary)  
- **REDIRECTED_FLIGHT** — flight rescheduled to a new destination at the exact same scheduled departure time (binary)  
- **ADVERSE_2H** — weather 2 hours prior to flight considered “adverse” (binary)

## Numerical

- **CRS_ELAPSED_TIME** — flight duration (minutes)  
- **DISTANCE** — straight-line route distance (miles)  
- **ELEVATION** — elevation of weather station (meters)  
- **RECENT_DEST_CANCELLED_COUNT_6_2H** — count of flights to same destination cancelled 2–8 hours before departure  
- **RECENT_DEST_DIVERSION_COUNT_6_2H** — count of flights to same destination diverted 2–8 hours before departure  
- **CRS_DEP_TIME_MINUTES** — scheduled departure time (minutes since midnight, 0–1440)  
- **CRS_DEP_TIME_MINUTES_NEXT** — scheduled departure time of next flight on tail (minutes since midnight, 0–1440), if found  
- **TIME_BETWEEN_ARR_AND_SCHEDULED_DEP** — minutes between previous arrival and current scheduled departure, if found  
- **TIME_BETWEEN_SCHEDULED_ARR_AND_SCHEDULED_NEXT_DEP** — minutes between current scheduled arrival and next scheduled departure on tail, if found  
- **PREV_DEP_DELAY** — departure delay of previous flight on tail, if found  
- **PREV2_ARRIVAL_DELAY** — arrival delay two flights back on tail, if found  
- **PREV2_DEP_DELAY** — departure delay two flights back on tail, if found  
- **PREV2_TIME_BETWEEN_ARR_AND_SCHEDULED_DEP** — minutes between arrival two flights back and scheduled departure one flight back  
- **UNIQUE_ORIGINS** — number of unique airports the plane has departed from to date  
- **NUM_WITHIN_100_MILES** — number of neighboring airports within 100 miles of destination  
- **USED_DAYS_SO_FAR** — number of days the plane has flown to date  
- **CALENDAR_DAYS_SO_FAR** — days since the plane’s first flight in the dataset  
- **TAIL_NUM_USE** — percent of days the plane has flown since first observed flight  
- **ORIGIN_RELATIVE_TRAFFIC** — flight count at destination on previous day divided by average flight count at the same airport over the previous calendar month

## Numerical — Graph

*Graphs were generated by inverting total flight counts between airports on a monthly basis. Features derived were applied to the next month’s flights to avoid leakage.*

- **ORIGIN_BETWEENNESS** — betweenness centrality of the origin airport  
- **ORIGIN_DEGREE** — degree centrality of the origin airport  
- **ORIGIN_CLOSENESS** — closeness centrality of the origin airport  
- **DEST_BETWEENNESS** — betweenness centrality of the destination airport  
- **DEST_DEGREE** — degree centrality of the destination airport  
- **DEST_CLOSENESS** — closeness centrality of the destination airport

## Numerical — Weather

*X ranges from integers 2–6 (inclusive), measured X hours prior to scheduled departure.*

- **HOURLYVISIBILITY_X** — visibility  
- **HOURLYWINDSPEED_X** — wind speed  
- **HOURLYWINDGUSTSPEED_X** — wind gust speed  
- **HOURLYPRECIPITATION_X** — precipitation  
- **HOURLYDRYBULBTEMPERATURE_X** — temperature  
- **HOURLYSEALEVELPRESSURE_X** — sea-level air pressure
